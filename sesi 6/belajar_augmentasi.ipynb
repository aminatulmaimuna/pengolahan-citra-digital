```python
#Melakukan Flip

from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open('/content/daun.jpg')

flip = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip()
])

fig, axes = plt.subplots(3, 3, figsize=(12, 12))
fig.subplots_adjust(wspace=0.4)
for ax in axes.flatten():
    ax.imshow(flip(img))
```

```python
#Melakukan Crop

import torchvision


img_crop = img
crop = torchvision.transforms.RandomCrop(size=340)

fig, axes = plt.subplots(3, 3, figsize=(7, 8))
fig.subplots_adjust(wspace=0.4)
for ax in axes.flatten():
    ax.imshow(crop(img_crop))
```

```python
# Random Rotation

img_rotation = img
rotate = torchvision.transforms.RandomRotation(degrees=20)

fig, axes = plt.subplots(3, 3, figsize=(8, 7))
fig.subplots_adjust(wspace=0.3)
for ax in axes.flatten():
  ax.imshow(rotate(img_rotation))
```

```python
#Random Resize Crop

img_resize_crop = img
resize_crop = torchvision.transforms.RandomResizedCrop(size=150)

fig, axes = plt.subplots(3, 3, figsize=(8, 8))
fig.subplots_adjust(wspace=0.3)
for ax in axes.flatten():
  ax.imshow(resize_crop(img))
```

```python
#center crop

img_center_crop = img

center_crop = torchvision.transforms.CenterCrop(size=500)
fig, axes = plt.subplots(3, 3, figsize=(8, 8))
fig.subplots_adjust(wspace=0.3)
for ax in axes.flatten():
  ax.imshow(center_crop(img_center_crop))

```

```python
#Random afiine translate

img_translate = img
translate = torchvision.transforms.RandomAffine(degrees=0, translate=(0.3, 0.3))

fig, axes = plt.subplots(3, 3, figsize=(8, 7))
fig.subplots_adjust(wspace=0.3)
for ax in axes.flatten():
  ax.imshow(translate(img_translate))
```

```python
#image afine shear

img_shear = img
shear = torchvision.transforms.RandomAffine(0, shear=20)

fig, axes = plt.subplots(3, 3, figsize=(8, 7))
fig.subplots_adjust(wspace=0.3)
for ax in axes.flatten():
  ax.imshow(shear(img_shear))
```
