```python
import cv2
from matplotlib import pyplot as plt

# Membaca gambar dalam mode grayscale
img = cv2.imread('miaw.jpg')

# Tampilkan gambar
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.title('RGB COLOR')
plt.imshow(img_rgb)
```

```python
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

plt.title('GRAY COLOR')
plt.imshow(img_gray, cmap='gray')
```

Membuat Histogram dari Citra Grayscale

```python
# Hitung histogram gambar grayscale
hist = cv2.calcHist([img_gray], [0], None, [256], [0,256])

# Plot histogram
plt.plot(hist, color='black')
plt.title("Histogram Grayscale")
plt.xlabel("Intensitas (0 - 255)")
plt.ylabel("Jumlah Piksel")
plt.show()
```

**Keteranngan : **

cv2.calcHist([img], [0], None, [256], [0,256]) artinya:

[img] → gambar input.

[0] → channel (0 untuk grayscale).

None → tanpa mask (menghitung seluruh gambar).

[256] → jumlah bin (0–255).

[0,256] → rentang nilai intensitas piksel.

plt.plot(hist) untuk menggambar histogram.

```python
# Hitung histogram untuk channel Red (index 2)
hist_red = cv2.calcHist([img_rgb], [2], None, [256], [0,256])

# Plot histogram Red
plt.plot(hist_red, color='r')
plt.title("Histogram Red Channel")
plt.xlabel("Intensitas (0-255)")
plt.ylabel("Jumlah Piksel")
plt.show()
```

```python
# Hitung histogram untuk channel green (index 1)
hist_green = cv2.calcHist([img_rgb], [1], None, [256], [0,256])

# Plot histogram Red
plt.plot(hist_green, color='g')
plt.title("Histogram Green Channel")
plt.xlabel("Intensitas (0-255)")
plt.ylabel("Jumlah Piksel")
plt.show()
```

```python
# Hitung histogram untuk channel green (index 0)
hist_blue = cv2.calcHist([img_rgb], [0], None, [256], [0,256])

# Plot histogram Red
plt.plot(hist_blue, color='b')
plt.title("Histogram Blue Channel")
plt.xlabel("Intensitas (0-255)")
plt.ylabel("Jumlah Piksel")
plt.show()
```

```python
# Warna channel
color = ('b', 'g', 'r')

plt.figure(figsize=(8,6))
for i, col in enumerate(color):
    hist = cv2.calcHist([img_rgb], [i], None, [256], [0,256])
    plt.plot(hist, color=col, label=col.upper())
    plt.xlim([0,256])

plt.title("Histogram Gabungan RGB")
plt.xlabel("Intensitas (0-255)")
plt.ylabel("Jumlah Piksel")
plt.legend()
plt.show()
```

```python
# Tambah kecerahan (misal +25)
img_bright = cv2.convertScaleAbs(img, alpha=1, beta=25)

# Tampilkan sebelum & sesudah
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Asli")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(img_bright, cv2.COLOR_BGR2RGB))
plt.title("Lebih Cerah (+100)")
plt.axis("off")

plt.show()
```

```python
# Warna channel
color = ('b', 'g', 'r')

plt.figure(figsize=(8,6))
for i, col in enumerate(color):
    hist = cv2.calcHist([img_bright], [i], None, [256], [0,256])
    plt.plot(hist, color=col, label=col.upper())
    plt.xlim([0,256])

plt.title("Histogram Gabungan RGB")
plt.xlabel("Intensitas (0-255)")
plt.ylabel("Jumlah Piksel")
plt.legend()
plt.show()
```
